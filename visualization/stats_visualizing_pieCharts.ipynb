First of all as always we do I wrote and import libraries which I needed in this chapter

```python
import matplotlib.pyplot as plt
import numpy as np
```

Here I created data for the plot

```python
nbins = 5
totalN = 100

rawdata = np.ceil(np.logspace(np.log10(1/2),np.log10(nbins-.01),totalN))
```

We have to prepare our data for pie chart

```python
uniquenums = np.unique(rawdata)
data4pie = np.zeros(len(uniquenums))

for i in range(len(uniquenums)):
    data4pie[i] = sum(rawdata==uniquenums[i])
```

To show the pie chart which we did some works to make it ready to be shown we can use this simple code 

```python
plt.pie(data4pie,labels=100*data4pie/sum(data4pie))
plt.show()
```

On the other hand we can do it like this:

```python
plt.pie(data4pie,labels=['zero','one','two','three','four'],explode=[0,.1,0,.15,0])
plt.show()
```

The code and the resualt which is our pie chart, made a different chart and in my opinion this one is much way better than the other one
